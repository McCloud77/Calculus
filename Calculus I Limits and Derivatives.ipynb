### 1. Introduction to limits.

The plot below shows that the limit of the function equals -2 as x approaches 1 from the left whereas the limit equals 2 as x approaches 1 from the right. Note that in this particular case, the function is not defined at x equals 1. Here, the value -2 represents the left-hand limit while the value 2 represents the right-hand limit. Because the left-hand limit is not equal to the right-hand limit, "the" limit does not exist (DNE).

```python
def f_x(x):
    return np.sqrt(3 - 5*x + x**2 + x**3) / (x - 1)

dmn_left = np.linspace(-1, 1, 100, endpoint=False)
dmn_right = np.linspace(3, 1, 100, endpoint=False)
left = [f_x(x) for x in dmn_left]
right = [f_x(x) for x in dmn_right]
plt.plot(dmn_left, left, color="blue", label="from the left")
plt.plot(dmn_right, right, color="blue", label="from the right")
plt.plot([1], [2], marker="o", color="blue", mfc="none", markersize=6)
plt.plot([1], [-2], marker="o", color="blue", mfc="none", markersize=6)
```

When the left-hand limit is equal to the right-hand limit, "the" limit exists and is equal to the left-hand limit (and, of course, the right-hand limit). In the example below, "the" limit equals 0.5. Note that for purposes of the limit, it does not matter if g(0) had an actual value, e.g., 0.01 or happens to be equal to the value of the limit.

```python
def g_x(x):
    return x / np.tan(2*x)

dmn_left = np.linspace(-1, 0, 100, endpoint=False)
dmn_right = np.linspace(1, 0, 100, endpoint=False)
left = [g_x(x) for x in dmn_left]
right = [g_x(x) for x in dmn_right]
plt.plot(dmn_left, left, color="blue", label="from the left")
plt.plot(dmn_right, right, color="blue", label="from the right")
plt.plot([0], [0.5], marker="o", color="blue", mfc="none", markersize=6)
```

When either the left-hand limit or the right-hand limit exists, but not both, "the" limit does not exist (DNE). This is shown in the example below where the left-hand limit equals 0 as x approaches 0 from the left whereas the function value goes to infinity as x approaches 0 from the right.

```python
def h_x(x):
    return (abs(x) + np.sin(x)) / x**2

dmn_left = np.linspace(-1, 0, 100, endpoint=False)
dmn_right = np.linspace(1, 0, 100, endpoint=False)
left = [h_x(x) for x in dmn_left]
right = [h_x(x) for x in dmn_right]
plt.plot(dmn_left, left, color="blue", label="from the left")
plt.plot(dmn_right, right, color="blue", label="from the right")
plt.plot([0], [0], marker="o", color="blue", mfc="none", markersize=6)
```

In the example below, the limit does not exist. As x approaches 0 from either the left or the right, the function value does not approach a finite number and also does not go to either negative or positive infinity. In fact, the function value oscillates between -1 and +1.

```python
def j_x(x):
    return np.sin(13/x)

dmn_left = np.linspace(-1, 0, 100, endpoint=False)
dmn_right = np.linspace(1, 0, 100, endpoint=False)
left = [j_x(x) for x in dmn_left]
right = [j_x(x) for x in dmn_right]
plt.figure(figsize=(15,5))
plt.plot(dmn_left, left, color="blue", label="from the left")
plt.plot(dmn_right, right, color="blue", label="from the right")
plt.axvline(x=0, linestyle="--")
```

In summary, the following limit behaviors are recognized.  
* The right-hand and left-hand limits may both exist and be equal.  
* The right-hand and left-hand limits may both exist, but may fail to be equal.  
* A right- and/or left-hand limit could fail to exist due to blowing up to positive or negative infinity. In this case the limit does not exist since infinity is not a real number.  
* A right- and/or left-hand limit could fail to exist because it oscillates between many values and never settles down. In this case the limit does not exist.  

The limit definition is:  

If a function f(x) approaches some value L as x approaches a from both the right and the left, then the limit of f(x) exists and equals L.

## Limit Laws.

Suppose the the limit of f(x) as x approaches a equals L and the limit of g(x) as x approaches a equals M, then:  
1. The limit of f(x) + g(x) as x approaches a equals L + M (Limit Law for Addition).  
2. The limit of f(x) - g(x) as x approaches a equals L - M (Limit Law for Subtraction).  
3. The limit of f(x) $*$ g(x) as x approaches a equals L $*$ M (Limit Law for Multiplication).
4. The limit of f(x) / g(x) as x appraoches a equals L / M if M is not 0 (Limit Law for Division).

### 2. Continuity.

Definition: A function f is continuous at x equals a if the limit of f(x) as x approaches a equals f(a).  

In the example below, the overall limit of f(x) as x approaches 0 does exist for the "blue" function (i.e., the left- and right-hand limits agree), but the overall limit does not equal f(0). It is said that f has a **removable discontinuity** at x = 0.  

If the left-hand limit of f(x) as x approaches a and the right-hand limit of f(x) as x approaches a both exist at a point x equals a, but they are **not** equal, then it is said that f has a **jump discontinuity** at x = a.  

In other words, a function f(x) is continuous if for every point ***c*** in the domain of f(x), the function f is continuous at the point x = c.

```python
def f_x1(x):
    return x**2 - x + 5
def f_x2(x):
    return x**2 + 2

dmn_left = np.linspace(-1, 0, 100, endpoint=False)
dmn_right = np.linspace(1, 0, 100, endpoint=False)
dmn_cont = np.linspace(-1, 1, 100)
left = [f_x1(x) for x in dmn_left]
right = [f_x1(x) for x in dmn_right]
f_cont = [f_x2(x) for x in dmn_cont]
plt.plot(dmn_left, left, color="blue", label="not continuous")
plt.plot(dmn_right, right, color="blue")
plt.plot(dmn_cont, f_cont, color="orange", label="continuous")
plt.plot([0], [5], marker="o", color="blue", mfc="none", markersize=6)
plt.legend(loc="upper right")
```

In the example below the functions sin(x) and cos(x) are continuous everywhere but the third function, tan(x), which is composed of the two functions sin(x) and cos(x), is not defined when cos(x) equals 0 (at any x which is multiple of 1/2$*$pi) and thus, tan(x) is not continuous everywhere. Nevertheless, tan(x) is continuous on the interval [0, 1/4$*$pi], for example.  

In general, when f(x) and g(x) are continuous everywhere then the composition of f(x) and g(x), f o g, is continuous everywhere.

```python
x = np.linspace(0, np.pi, 360)

y1 = np.sin(np.pi * x)
y2 = np.cos(np.pi * x)
y3 = np.tan(np.pi * x)

f = plt.figure(figsize=(12,5))
ax = f.add_subplot(131)
ax2 = f.add_subplot(132)
ax3 = f.add_subplot(133)
ax.plot(x,y1)
ax2.plot(x,y2)
ax3.plot(x,y3)
```

The **Intermediate Value Theorem** states that if f is a function which is continuous (!) on the interval [a, b], and M lies between the values of f(a) and f(b), then there is at least one point c between a and b such that f(c) = M.  

This theorem is often referred to as a deep result because it takes information that is local by nature, and allows you to conclude a global result.  

In the example below, the Intermediate Value Theorem implies that there must be some c between x = 0 and x = 9 for the first function since this function is continuous on [0, 9]. In contrast, since the second function is not continuous on the interval [-1, 1], the Intermediate Value Theorem does not imply that there must be some c between x = -1 and x = 1 where the function is zero (has a root).

```python
x1 = np.linspace(0, 9)
x2_a = np.linspace(-1, 0, endpoint=False)
x2_b = np.linspace(0, 1, endpoint=False)
x2 = np.append(x2_a, x2_b)
#x2 = np.linspace(1, 0, endpoint=False)
y1 = np.sqrt(x1) - np.cos(x1)
y2 = 4*x2**-3 - x2 - np.sin(np.pi*x2)

plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.subplot(1,2,2)
plt.plot(x2, y2)
```

### 3. Limits of quotients.

Given a function f(x) which is equal to L when x approaches a and a function g(x) which is equal to M when x approaches a, we can distinguish three possibilities for the limit of f(x) / g(x) when x approaches a.  

1. If M does not equal 0, then the limit of f(x) / g(x) equals L / M as x approaches a.
2. If M equals 0 and L does not equal 0, the limit of f(x) / g(x) DNE as x approaches a.
3. If M equals 0 **and** L equals 0, the limit of f(x) / g(x) may exist as x approaches a.  

In fact, when both L and M equal 0, the limit could be anything. Remember that limits as x approaches 0 only concern themselves with values of x that are ***near*** o, but ***no equal*** to 0. For such values, the numerator and denominator will be close to zero, but not equal to it.

The example below shows what might happen.

```python
print("Really small divided by something small yields really small:\t\t\t", 0.00000013 / 0.0025)
print("Small divided by something 'equally' small yields neither big nor small:\t", 0.000033 / 0.000029)
print("Small divided by something really small yields something big:\t\t\t", 0.0094 / 0.00000023)
print("Small divided by something smaller yields neither big nor small:\t\t", 0.00023 / 0.000081)
```

The plot below shows a function which is not defined at x = 1 and x = 2. Nevertheless, the limit as x approaches 1 can be calculated by factoring the function to (x - 1) $*$ (x + 3) / (x - 1) $*$ (x - 2) which simplifies to (x + 3) / (x - 2). Substituting x with 1, we see that the numerator goes to 4 and the denominator goes to -1 as x approaches 1. In contrast, the limit of the function as x approaches 2 from the left does not exist (DNE) and, in particular, the function blows up to negative infinity since substituting a value just a tiny bit smaller than 2 into the simplified expression (x + 3) / (x - 2) leads to the division with a really small negative number.

```python
x = np.linspace(0.5, 2, 1000, endpoint=False)
y = (x**2 + 2*x - 3) / (x**2 - 3*x + 2)
plt.plot(x,y)
plt.plot([1], [-4], marker="o", color="red", mfc="none", markersize=8)
plt.show()
```

### 4. The Derivative.

If the ***average*** velocity of a vehicle between 8:00 and time b is given by the formula:

\begin{equation*}
\frac{f(b) - f(8)}{b - 8}
\end{equation*}

we can take the limit as b approaches 8:00 in order to obtain the ***instantaneous*** velocity at 8:00 exactly.  

We say that **derivative** of f(x) at x = a equals the instantaneous rate of change of f(x) at x = a.  

\begin{equation*}
f'(a) = 
\frac{lim}{b -> a}
\frac{f(b) - f(a)}{b - a}
\end{equation*}

Suppose we are throwing a pumpkin from the top of a building which is 100 meters high and the initial direction is up, then a possible equation representing the velocity of the pumpkin at a given point is given by:

\begin{equation*}
 v = 100 + 20t - 5t^2
\end{equation*}

```python
def throw_pumpkin(upper_bound):
    t = np.linspace(0, upper_bound)
    y = [(100 + 20*x -5*x**2) for x in t]
    return (t, y)

upper_bound = 7
x, y = throw_pumpkin(upper_bound)
plt.plot(x, y)
plt.axvline(x=2, linestyle="--", color='red')
```

Note that in the example shown above the function is increasing until t = 2 after which it decreases. At t = 2 the function is neither increasing or decreasing. Thus, checking with some values for t we see that:  

* At t < 2, the function is increasing **and** the derivative of the function is greater than zero.
* At t = 2, the function is neither increasing or decreasing **and** the derivative of the function is zero.
* At t > 2, the function is decreasing **and** the derivative of the function is smaller than zero.

Q: A truck is going to make a trip of 500 km. If it is loaded with  pounds of cargo, its fuel efficiency on the trip will be f(x) miles per gallon. What units is f'(5000) measured in?  

A: The derivative of any function is always measured in units of the output of the function, divided by units of the input of the function. The output here is measured in miles per gallon (mi/g), and the input is measured in pounds (p), so the derivative is measured in miles per gallon per pound: (mi/g)/p, or mi/(g$*$p).

Q: Do you expect f'(5000) to be positive, negative, or zero?  

A: Negative. If the weight of the cargo increases, the fuel efficiency of the truck will decrease. This means that f'(5000) should be negative.

If the derivative of f(x) exists at x = a, then the tangent line exists. The tangent line may exist if the derivative is undefined at x = a though.

### 5. The Derivative as a Function.

The Power Rule:

If f(x) = x^n, what is f'(x)?

Per the definition of the derivative:
\begin{equation*}
f'(x) = \frac{lim}{b -> x} \frac{f(b) - f(x)}{b - x} = \frac{lim}{b -> x} \frac{b^n - x^n}{b - x}
\end{equation*}
and because:
\begin{equation*}{b^n - x^n} = {(b - x)(b^{n-1} + b^{n-2}x + b^{n-3}x^2 + .... + bx^{n-2} + x^{n-1})}
\end{equation*}
This turns into:

\begin{equation*}
\frac{lim}{b -> x} {(b^{n-1} + b^{n-2}x + b^{n-3}x^2 + .... + bx^{n-2} + x^{n-1})}
\end{equation*}
Since this is a polynomial we can fill in x for b:

\begin{equation*}
x^{n-1} + x^{n-2}x + x^{n-3}x^2 + .... + xx^{n-2} + x^{n-1}
\end{equation*}

Since all terms (n in total) evaluate to x^(n-1) the derivative is:

\begin{equation*} {nx^{n-1}} \end{equation*}

It would be really nice if we had a (set of) functions which could return the derivative and the value of the derivative at a specific value of x without having to modify functions etc. The functions below do just that.  

The function ***split_poly*** returns an array with the components of a polynomial (!) function.  

The function ***deriv*** takes a list containing each component of a polynomial and returns the derivative both as a string and an array with the components of the derivative.  

The function ***calc_deriv*** calculates the value of the derivative based on a specific value of x.

```python
#input_poly = input('please enter the polynomial as a string in the form '3x**3 + -5x**4 + 30'
poly = '3x**3 + -5x**4 + 30'
def split_poly(B):
    """
    Split a polynomial into its components.
    """
    return [x.strip() for x in B.split('+')]

res1 = split_poly(poly)
print(res1)
```

```python
def deriv(B):
    B_drv = []
    for parameter in B :
        if 'x' in parameter :
            index_x = parameter.index('x')
            multiply = parameter[0:index_x]
            # multiply referes to the number multiplied by 'x' , which is '9' in '9x**2'
            if multiply == '' :
                multiply = '1' #in case the user wrote 'x**3' instead of '1x**3'
            power = parameter[index_x+3:]
            # power referes to the power of 'x' , which is '2' in '9x**2'
            if power != '0' :
                drv_parameter = str(float(multiply) * float(power)) + 'x**' + str(float(power)-1) 
            else :
                drv_parameter = str(0) #in case the user wrote a parameter '4x**0' instead of '4'
            B_drv.append(drv_parameter) #filling the list
        else: #if there is no 'x' in the parameter
            drv_parameter = str(0)
            B_drv.append(drv_parameter) # for use in the function calc_deriv
    result = ' + '.join(B_drv) # for display
    return (result, B_drv)

res2, derivative = deriv(res1)
print("The derivative is:", res2)
```

```python
def calc_deriv(B_drv, val):
    x_value = val
    for parameter in B_drv :
        index_parameter = B_drv.index(parameter)
        B_drv[index_parameter] = parameter.replace('x' , '*' + x_value ) # to replace '4x**3' with '4*number**3'
        B_drv[index_parameter] = str(eval(B_drv[index_parameter]))
    result = eval(' + '.join(B_drv))
    return(result)

val='2'
print("The derivative is", calc_deriv(derivative, val), "at x =", float(val))
```

### 6. Leibniz notation.

The Leibmiz notation is particularly useful when we are modeling a system and would like to keep track of the relationship of the variables involved as we are taking derivatives. Instead of f'(x) we now write dy/dx to indicate the change in y as a result of the change in y. Let's look at an example to clarify the usefulness of Leibniz' notation.

Recall the formulas for the area of a circle: \begin{equation*}A = {\pi}r^2 \end{equation*}

and for the circumference of a circle: \begin{equation*}c = {2\pi}r \end{equation*}

Expressing the area A in terms of c: \begin{equation*}A = \frac{c^2}{4\pi}\end{equation*}

Suppose we would like to determine how the area of a circle changes with respect to the circumference. The two derivatives we need are:  

\begin{equation*}\frac{dA}{dr} = {2\pi}r\end{equation*}

\begin{equation*}\frac{dA}{dc} = \frac{c}{2\pi}\end{equation*}

### 7. Second derivatives and higher.

The second derivative of a function $f(x)$ is the first derivative of $f^{'}(x)$ and is denoted by $f^{''}(x)$ or $\frac{d^{2}f}{dx^{2}}$. This can be extended to higher derivatives (if they exist).

Let $f(x) = sin(x)$ for $x$ $\epsilon$ $[-\pi, \pi]$ (see below). It should be easy to see that the derivative of this function at $x$ $=$ $-3$ is negative (the function is decreasing). When we evaluate the first derivative from $-2$ to $0$, the first derivative (itself a function) goes from being negative to positive. When we consider the value of the ***second*** derivative, we see that the value (the slope of the first derivative) is initially negative but is continuously increasing over the interval $[-2, 0]$. This is referred to as "concave up".  

Points where the graph of a function changes from concave up to concave down, or vice versa, are called inflection points. 

```python
x_coord = np.linspace(-np.pi, np.pi)
y = [np.sin(x) for x in x_coord]
plt.plot(x_coord, y)
```
